# Introduction to the Business Problem:

Our client, John Smith, from Europe wants to open a location of his successful authentic Italian food restaurant in North America. This will be his first time venturing in to the North American food restaurant market and he would like to open it first either in New York or Toronto.  He would like us to recommend him the ideal venue for his restaurant by leveraging our expertise and knowledge.

New York and Toronto are both the largest cities of their respective countries, where New York is the financial hub of the United States of America and similarly Toronto is the financial hub of Canada. Both the cities have high population density, higher cost of living, and varied ethnicities and lifestyle. The objective of this project is to find the similarities and differences between New York and Toronto based on their individual neighbourhoods and popularity of Italian restaurants to identify and recommend the most suitable venue for John Smith's signature Italian restaurant.
